# **Operadores Genéticos**
---

- **Cruzamiento**

Se toman dos padres (individuos de la población) y se recombinan sus genes para generar dos nuevos individuos hijos, el offspring o la descendencia formará la siguiente generación de individuos, una nueva población para comenzar a iterar sobre las generaciónes

cruza.png

- **Mutación**

Se toma la población y de manera aleatoria se toma una determinada cantidad de posibles individuos mutantes para tomar un de sus genes y modificarlo con una probabilidad del $5\%$, que es la tasa de mutación observada en las especies. Los mutantes se dejan dentro de la población.

mutacion.png

- **Selección**

Para evaluar la aptitud de los individuos de una población, se toman los $n$ mejores individuos de la misma según algún criterio establecido para el problema y se guardan en un "**Salón de la Fama**", el resto de la población se elimina para dejar solamente a los mejores. A esta población mejorada se le aplica el operador de cruzamiento para rellenar la población con el número orignal de individuos.

Existen varios mecanismos de selección:
- Por **Ruleta**
- Por **Torneo**
- Por **Rango**

Los operadores de cruzamiento y mutación buscan emular los **mecanismos genéticos**, mientras que el operador fitness se basa en los **mecanismos competitivos**.

# **ALGORITMOS GENÉTICOS CUÁNTICOS**

Con la intención de unir el cómputo cuántico y los algoritmos genéticos, un nuevo algoritmo evolutivo surgió. Los algoritmos genéticos cuánticos (QGAs) están inspirados en la mecánica cuántica. A diferencia de los GAs clásicos que tratan con bits, los QGAs manipulan quantum bits (qubits)

- **Qubit**: unidad más pequeña de información que puede ser almacenada en dos estados cuánticos o en una superposición de ambos.

\begin{equation}|\Psi > = \alpha |0> + \beta |1> \end{equation}

donde

\begin{equation} |0>=(1,0)^T \ \ |1>=(0,1)^T \end{equation}

Una representación geométrica de los qubits es através de la esfera de Bloch

esfera.png

Con esta nueva idea y metodología se construyen las poblaciones que son conocidas como poblaciones cuánticas.

\begin{equation} Q(t)=
\begin{bmatrix}
\alpha_1^t & \alpha_2^t & \cdots & \alpha_m^t \\
\beta_1^t & \beta_2^t & \cdots & \beta_m^t
\end{bmatrix}
\end{equation}

Cuando se inicializa una población cuántica, las amplitudes de probabilidad a $t=0$ serán $\frac{1}{\sqrt{2}}$, esto con la finalidad de que todos los estados posibles tengan la misma probabilidad.

En la mecánica cuántica, al querer evolucionar un estado inicial a uno final se debe hacer uso de la ecuación de Schrödinger

\begin{equation} i\hbar \frac{\partial}{\partial t}|\Psi(t)> = H(t) |\Psi(t)> \end{equation}

Supongamos que $U(t)$ es una solución de la ecuación de Schrödinger, de tal manera que dado un estado inicial $|\Psi(0)>$ la solución a un estado $t$ vendrá dado como:

\begin{equation} |\Psi(t)>=U(t)|\Psi(0)> \end{equation}

Interpretemos esto como:
- $|\Psi(0)> \Longrightarrow$ entrada o población inicial
- $|\Psi(t)> \Longrightarrow$ salida (evolución+medición)
- $U \Longrightarrow$ operador de evolución unitario

En los QGAs al operador $U$ se le conoce como quantum gate (**Q-gate**)



# **Operadores Genéticos Cuánticos**

- **Rotation Gate**

\begin{equation} \mathbf{Q}_i^{t+1} = U(t)\cdot \mathbf{Q}_j^t \end{equation}

donde:

\begin{equation}
U(t)=
\begin{pmatrix}
\cos(\delta\theta_j) & -\sin(\delta\theta_j) \\
\sin(\delta\theta_j) & \cos(\delta\theta_j)
\end{pmatrix}
\end{equation}

Este operador busca ampliar o disminuir las amplitudes de probabilidad de los qubits en correlación con el cromosoma con el máximo valor fitness.

rotacion.png

- **Quantum Mutation Gate** \\
El proceso consiste en intercambiar las amplitudes del j-ésimo qubit mediante la matriz de Pauli $\sigma_x$:

\begin{equation}
\begin{pmatrix}
\beta_j^{t+1} \\
\alpha_j^{t+1}
\end{pmatrix}
=
\begin{pmatrix}
0 & 1 \\
1 & 0
\end{pmatrix}
\begin{pmatrix}
\alpha_j^{t} \\
\beta_j^{t}
\end{pmatrix}
\end{equation}

- **Quantum crossover gate** \\
Es el análogo cuántico al operador de cruza clásico, sólo que ahora se está operando sobre amplitudes de probabilidad.

# **EJEMPLO DE APLICACIÓN QGA**

A continuación aplicaremos los conceptos de los algoritmos genético cuánticos para construir un código que determine el minimo de la función:
\begin{equation} f(x) = \left|\frac{x-5}{2+\sin(x)}\right| \end{equation}

Comencemos graficando la función para ver su comportamiento.


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-50,50,100)
y = abs((x-5)/(2+np.sin(x)))
plt.rcParams.update({'font.size': 10})
plt.plot(x,y,'r')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

Podemos notar que la función tiene un mínimo entre 0 y 10

```python
#Definamos los parámetros que empleará el algoritmo:

num_qubits = 10
tam_pob = 5 #Número de soluciones en la población
generaciones = 10
tasa_mutacion = 0.01 #Probabilidad de mutacion
theta = np.pi/50 #Ángulo de rotación para el Rotation Gate
limite_inferior = -10.
limite_superior = 10.

```

Cualquier algoritmo genetico requiere de una función objetivo a optimizar definamosla

```python
def fun_objetivo(x):
  return abs((x-5)/(2+np.sin(x)))
```

Definamos la funcion que inicialice la población de posibles soluciones.
Para tener una misma amplitud de probabilidades inicialicemos con $\frac{1}{\sqrt{2}}$

```python
def inicializar_poblacion():
    return np.full((tam_pob, num_qubits), 1/np.sqrt(2))
```

A continuación creemos una función que reciba como argumento la población (P) de soluciones (matriz de qubits) y creando una matriz de números aleatorios (random_matriz) entre 0 y 1 vamos a comparar con la matriz P con random_matriz para establecer si el qubit esta en el estado 0 o 1.

```python
# Generación de soluciones a partir de las probabilidades de los qubits
def generar_soluciones(Poblacion):
    random_matriz = np.random.rand(tam_pob, num_qubits)
    binary_matriz = (random_matriz < Poblacion).astype(int)
    soluciones = ["".join(map(str, binary_matriz[i])) for i in range(tam_pob)]
    return soluciones
```

Como ya se han estado creando cadenas binarias es importante convertir esos numeros binarios a decimales, para esto definamos la siguiente función:

```python
def binario_a_real(cadena_binaria, limite_inferior, limite_superior):
    valor_decimal = int(cadena_binaria, 2)
    max_decimal = 2**num_qubits - 1
    return limite_inferior + (limite_superior - limite_inferior) * valor_decimal / max_decimal
```

Ahora definamos una función que evalúe la población de soluciones binarias transformandolas en valores reales y mediante el uso de la función objetivo se obtenga una evaluación de cada solución.

```python
def evaluar_poblacion(soluciones, limite_inferior, limite_superior):
    return [fun_objetivo(binario_a_real(sol, limite_inferior, limite_superior)) for sol in soluciones]

#La función devuelve una lista con los resultados, uno para cada solución binaria
```

Definamos una función que utilice el operador cuántico Rotation Gate para actualizar las poblaciones en el QGA

```python
def rotation_gate(P, soluciones, mejor_solucion):
    for i in range(tam_pob):
        for j in range(num_qubits):
            if soluciones[i][j] != mejor_solucion[j]:
                # Determinamos la dirección de rotación
                if mejor_solucion[j] == '1':
                    P[i, j] += np.sin(theta) * (1 - P[i, j])
                else:
                    P[i, j] -= np.sin(theta) * P[i, j]
    P = np.clip(P, 0.05, 0.95)  # Limitamos valores extremos
    return P
```

Ahora la función para el operador Quantum Mutation Gate

```python
def quantum_mutation_gate(P):
    mutation_mask = np.random.rand(tam_pob, num_qubits) < tasa_mutacion
    P = np.where(mutation_mask, 1 - P, P)
    return P
```

**Algoritmo Principal**

```python
def quantum_genetic_algorithm(limite_inferior, limite_superior):
    P = inicializar_poblacion()
    mejor_solucion = None
    mejor_fitness = float('inf')

    for gen in range(generaciones):
        soluciones = generar_soluciones(P)
        fitness = evaluar_poblacion(soluciones, limite_inferior, limite_superior)

        actual_mejor_fitness = min(fitness)
        actual_mejor_solucion = soluciones[np.argmin(fitness)]

        if actual_mejor_fitness < mejor_fitness:
            mejor_fitness = actual_mejor_fitness
            mejor_solucion = actual_mejor_solucion

        P = rotation_gate(P, soluciones, mejor_solucion)
        P = quantum_mutation_gate(P)

        if gen % 10 == 0:
            print(f"Generación {gen}, Mejor aptitud: {mejor_fitness}")

    # Convertir la mejor solución binaria a su valor real
    mejor_real_solucion = binario_a_real(mejor_solucion, limite_inferior, limite_superior)
    return mejor_real_solucion, mejor_fitness
```

Ejecutamos el algoritmo genético cuántico

```python
mejor_solucion, mejor_fitness = quantum_genetic_algorithm(limite_inferior, limite_superior)
print(f'Mejor solución: x = {mejor_solucion}, f(x) = {mejor_fitness}')
```
