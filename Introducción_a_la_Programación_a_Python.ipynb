##Matriz inversa
La matriz inversa tiene muchas propiedades interesantes, como:


*  Multiplicar la matriz inversa por la matriz original resulta en la matriz identidad: \\
$B^{-1} \times B = \mathbb{I}$ \\
  donde $\mathbb{I}$ es una matriz sólo con unos en su diagonal:
  \begin{equation}
  \mathbb{I}=
  \begin{pmatrix}
  1 & 0 \\
  0 & 1 \\
  \end{pmatrix}
   \space ó \space \space  
  \mathbb{I}=
  \begin{pmatrix}
  1 & 0 & 0 \\
  0 & 1 & 0 \\
  0 & 0 & 1
  \end{pmatrix}
  \end{equation}
*  Encontrar la matriz inversa nos permite resolver un sistema de ecuaciones como el siguiente

  $a_1 x + a_2 y + a_3 z=c$ \\
  $b_1 x+ b_2 y + b_3z=d$ \\
  $c_1 x + c_2 y + c_3 z =e$ \\

  Podemos expresarlo en un sistema matricial:
\begin{equation}
  \begin{pmatrix}
  a_1 & a_2 & a_3 \\
  b_1 & b_2 & b_3 \\
  c_1 & c_2 & c_3 \\
  \end{pmatrix}
  \times
  \begin{pmatrix}
  x \\ y \\ z
  \end{pmatrix}
  =
  \begin{pmatrix}
  c \\ d \\ e
  \end{pmatrix}
\end{equation}
  Esto es:
  \begin{equation}
  A \times \vec{x} = \vec{c}
  \end{equation}
  Si encontramos la matriz inversa, $A^{-1}$, y calculamos el producto matricial con $\vec{c}$ (el término de la derecha, llamado *vector de coeficientes*) el resultado será una matriz cuyas componentes son iguales a los valores que toman $x$, $y$ y $z$. \\
  Esto es, multiplicando por $A^{-1}$ la ecuación anterior:
  \begin{equation}
  A^{-1} \times A \times \vec{x} = A^{-1} \times \vec{c}
  \end{equation}

  \begin{equation}
  \mathbb{I} \times \vec{x} = A^{-1} \times \vec{c}
  \end{equation}
  Luego, la matriz identidad multiplicada por el vector $\vec{x}$ resulta en el mismo vector, y obtenemos:
  \begin{equation}
  \vec{x} = A^{-1} \times \vec{c}
  \end{equation}






###**Podemos aplicar estos conceptos y resolver fácilmente un sistema de ecuaciones con el uso de la librería Numpy**

Queremos resolver el siguiente sistema de ecuaciones: \\
\begin{equation}
x+y+z=4\\
x+2y+4z=0 \\
x+3y+9z=12
\end{equation}
Éste lo podemos representar como:
\begin{equation}
  \begin{pmatrix}
  1 & 1 & 1 \\
  1 & 2 & 4 \\
  1 & 3 & 9 \\
  \end{pmatrix}
  \times
  \begin{pmatrix}
  x\\ y \\ z
  \end{pmatrix}
  =
  \begin{pmatrix}
  4\\ 0 \\ 12
  \end{pmatrix}
\end{equation}

```python
A=np.array([[1,1,1],[1,2,4],[1,3,9]])
print("A=")
print(A)
c=np.array([[4],[0],[12]]) #Vector de coeficientes
x=linalg.solve(A,c) #Vector x, valores de las incógnitas
print("x=",x[0]," y=",x[1]," z=",x[2])
```

##**Teorema de la descomposición espectral**

Si A es una matriz cuadrada (de dimensiones n $\times$ n), y simétrica, se tienen los siguientes resultados:


1.   Los valores propios (eigenvalores o autovalores) son reales
2.   Los vectores propios (eigenvectores o autovectores) son ortogonales entre sí
3.   La matriz original A se puede diagonalizar, es decir, existe una matriz P que cumple que:
  $P^{-1} \times A \times P = D$ \\
  donde D es una matriz diagonal.

Ejemplos de matriz diagonal:
\begin{equation}
\begin{pmatrix}
i & 0 & 0 \\
0 & j & 0 \\
0 & 0 & k
\end{pmatrix}
\end{equation}
Finalmente, todo ello para *descomponer* la matriz A en su *espectro* (otras tantas matrices) que se forman con los valores y vectores propios de A.

```python
A=np.array([[7,-2],[-2,4]])
print("A=")
print(A)
eigenvalores,eigenvectores=linalg.eig(A)
# Separamos primero los eigenvalores
a=eigenvalores[0]
b=eigenvalores[1]
# Separamos los eigenvectores
x=eigenvectores[:,0]
y=eigenvectores[:,1]

print("eigenavalor 1 = ", a, ", eigenvalor 2 = ", b)
print("eigenvector 1=",x)
print("eigenvector 2=", y)

print("Producto punto =",np.vdot(x,y)) #Estos vectores son ortogonales entre sí, pues su producto punto es cero
```

# Oscilador Armónico Amortiguado (OAA)

El oscilador armónico amortiguado es un modelo matemático para describir el movimiento de objetos que oscilan respecto a un punto en el espacio y que eventualmente se detienen. El parámetro de amortiguamiento $\beta=b/2m$ en la ecuación
\begin{equation}
m\frac{d^2x}{dt^2}+b\frac{dx}{dt}+kx=0
\end{equation}
es característico de cada movimiento de este tipo. Se obtiene
\begin{equation}
\frac{d^2x}{dt^2}+2\beta\frac{dx}{dt}+w_o^2x=0
\end{equation}
donde $w_o^2=k/m$ es la frecuencia natural de oscilación del objeto.

La solución general de esta ecuación diferencial de movimiento es
\begin{equation}
x(t)=e^{-\beta t}\left[ A\text{exp}\left(t\sqrt{\beta^2-w_o^2}\right) + B\text{exp}\left(-t\sqrt{\beta^2-w_o^2}\right)  \right]
\end{equation}

Y según la relación entre el parámetro de amortiguamiento $\beta$ y la frecuencia natural de oscilación $w_o$ esta solución, y el fenómeno físico, caen en determinado caso. Analizaremos el caso de oscilaciones subamortiguadas donde $w_o^2>\beta^2$.

La solución, que es el modelo no lineal que ajustaremos a los datos, es
\begin{equation}
x(t)=Ae^{-\beta t} \cos{(wt-\delta)}
\end{equation}

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.optimize import curve_fit
```

```python
dataset = pd.read_csv('https://raw.githubusercontent.com/latorresl/PhyLab-Statistics/refs/heads/main/The%20Data%20Sets/OAAm.csv')
x = dataset['tOAAm'] # Posición en x
y = dataset['xOAAm'] # Posición en y
```

```python
# Definimos un modelo. Es una función, cuyo primer argumento son los valores
# del eje x y el resto de los argumentos son los parámetros a hallar

def OAA(x, a, b, c, d):
    return a * np.exp(-b*x) * np.cos(c * x + d)

# Parámetros iniciales con los que vamos a iniciar el proceso de ajuste (meras aproximaciones)
parametros_iniciales=[3, 0.25, 3, -np.pi]

# Hacemos el ajuste con curve_fit
popt, pcov = curve_fit(OAA, x, y, p0=parametros_iniciales)
incertidumbre = np.sqrt(np.diag(pcov))

a,b,c,d = popt

# curve_fit devuelve dos resultados. El primero (popt) son los
# parámetros óptimos hallados. El segundo (pcov) es la matriz de
# covarianza de los parametros hallados.

# Se genera el intervalo de graficado
x_modelo  = np.linspace(min(x), max(x), 1000)

# Se definen las funciones contenedoras
cont1 = a*np.exp(-b*x)
cont2 = -a*np.exp(-b*x)

# popt tiene los parámetros hallados en orden
print('Parámetros ordenados: {}'.format(popt))

# De la matriz de covarianza podemos obtener los valores de desviacion
# estandar de los parametros ubicados en su diagonal
pstd = np.sqrt(np.diag(pcov))

nombres_de_param=['a','b','c','d']
print('\nParámetros hallados:')
for i,param in enumerate(popt):
    print('{:s} = {} ± {}'.format( nombres_de_param[i] , param , pstd[i]/2) )

print('\nCoeficientes: {}\nIncertidumbre: {}\n\nMatriz de covarianza:\n {}'.format(popt,incertidumbre,pcov))

# Calcular R^2
y_mean = np.mean(y)
TSS = np.sum((y - y_mean)**2)
y_pred = OAA(x, a, b, c, d)
RSS = np.sum((y - y_pred)**2)
r_squared = 1 - (RSS / TSS)

print(f'\nCoeficiente R^2: {r_squared}')
```

### Covarianza y matriz de covarianza
La covarianza es una medida de cuánto varían de manera conjunta dos variables aleatorias.

\begin{equation}
\operatorname{Cov}(X_i, X_j) = \mathbb{E}[(X_i - \mathbb{E}[X_i])(X_j - \mathbb{E}[X_j])]
\end{equation}

De modo que la matriz de covarianza es

\begin{equation}
\Sigma = \begin{pmatrix}
\operatorname{Cov}(X_1, X_1) & \operatorname{Cov}(X_1, X_2) & \cdots & \operatorname{Cov}(X_1, X_n) \\
\operatorname{Cov}(X_2, X_1) & \operatorname{Cov}(X_2, X_2) & \cdots & \operatorname{Cov}(X_2, X_n) \\
\vdots & \vdots & \ddots & \vdots \\
\operatorname{Cov}(X_n, X_1) & \operatorname{Cov}(X_n, X_2) & \cdots & \operatorname{Cov}(X_n, X_n)
\end{pmatrix}
\end{equation}

La covarianza de dos variables aleatorias iguales es la varianza de esa variable, $\operatorname{Cov}(X_i, X_j)=\operatorname{Var}(X)=\sigma_X^2$.

```python
plt.figure()
plt.plot(x, y, 'o', label='Mediciones', color='blue', markersize=2.5, linewidth=10)
plt.plot(x, cont1, '--', label='Función contenedora', color='green')
plt.plot(x, cont2, '--', label='Función contenedora', color='purple')
plt.plot(x_modelo, OAA(x_modelo, *popt), 'r-', label='Oscilador Ajustado')
plt.title('Oscilador Armónico Amortiguado')
plt.legend(loc='upper right', fancybox = False, edgecolor = 'black', fontsize = 10)
plt.grid()
plt.xlabel('Tiempo [$s$]')
plt.ylabel('x [$m$]')
plt.tight_layout()
```

# Catástrofe Ultravioleta



Cuando la teoría de la electrodinámica clásica falló al explicar la emisión electromagnética de un cuerpo en equilibrio térmico con el ambiente con la ecuación de Rayleigh-Jeans
\begin{equation}
I(ν)=\frac{8π}{c^3}kTν^2
\end{equation}
que fallaba al describir los resultados experimentales sobre la intensidad de la radiación para una frecuencia $ν$, Max Planck propuso que esta intensidad estaba descrita por la ecuación
\begin{equation}
I(ν)=\frac{8πh}{c^3}\frac{ν^3}{e^{hν/kT}-1}
\end{equation}
Donde introdujo el concepto de cuanto. Si la energía de la radiación en la ecuación anterior es
\begin{equation}
ϵ=\frac{hν}{e^{hν/kT}-1}
\end{equation}
el cuanto de energía para una radiación de frecuencia $ν$ es $ϵ=hν$. Graficaremos el comportamiento de ambos modelos, notando que el modelo de Planck se ajustó a los resultados de los experimentos realizados en esa época.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
h = 6.626e-34  # Constante de Planck (J·s)
c = 3.0e8      # Velocidad de la luz (m/s)
k = 1.38e-23   # Constante de Boltzmann (J/K)
pi = np.pi

T = 1500

# Rango de frecuencias
f = np.linspace(1e10, 1e15, 1000)
```

```python
# Ecuación de Rayleigh-Jeans
def rayleigh_jeans(f, T):
    return (8 * pi * k * T * f**2) / (c**3)

# Ecuación de Planck
def planck(f, T):
    return (8 * pi * h * f**3) / (c**3 * (np.exp(h * f / (k * T)) - 1))

# Calcular las densidades de energía espectral
G_f_rayleigh = rayleigh_jeans(f, T)
u_f_planck = planck(f, T)
```

```python
plt.rcParams['font.family'] = 'serif'
font = {'family': 'serif', 'color': 'black', 'weight': 'light', 'size': 14}

plt.figure(figsize=(8, 6))
plt.plot(f, G_f_rayleigh, label="Rayleigh-Jeans", color='deeppink', linestyle='--', linewidth=4)
plt.plot(f, u_f_planck, label="Planck", color='deeppink', linewidth=4)
plt.xlabel("Frequency, v (Hz)")
plt.ylabel("Spectral energy density, u(v)dv")
plt.xlim(0, 4e14)  # Limitar el rango de frecuencias
plt.ylim(0, max(u_f_planck) * 2)  # Limitar el rango de la densidad de energía
plt.grid(True)

# Ajuste de escala en el eje x para múltiplos de 10^14
plt.gca().set_yticklabels([])
xticks = np.arange(0, 5e14, 1e14)  # Marcas en múltiplos de 10^14
plt.xticks(xticks)  # Aplicar marcas en el eje x
plt.gca().ticklabel_format(style='sci', axis='x', scilimits=(14, 14))

plt.gca().set_facecolor('pink')
plt.grid(color='white', linewidth=1.5, linestyle='-')

y_center = max(u_f_planck)
plt.text(1.1e14, y_center * 1.5, 'Rayleigh-Jeans', fontsize=15, ha='center')
plt.text(3e14, y_center * 0.2, 'M. Planck', fontsize=15, ha='center')
plt.show()
```

# Filtro Pasa Bajas

Un filtro pasa bajas permite el paso de frecuencias bajas y atenúa (reduce) las frecuencias altas a partir de una frecuencia de corte específica ($f_c$). Es común en aplicaciones donde solo se requieren frecuencias bajas, como en la eliminación de ruido de alta frecuencia.

La frecuencia de corte se calcula como:
\begin{equation}
f_c = \frac{1}{2 \pi RC}
\end{equation}
donde R es la resistencia y C es la capacitancia. Por debajo de $f_c$, la señal pasa con mínima atenuación, y por encima de $f_c$, se atenúa cada vez más.

# Filtro Pasa Altas

Un filtro pasa altas permite el paso de frecuencias altas y atenúa las frecuencias bajas. Es útil en aplicaciones donde solo se quieren frecuencias altas, como en la eliminación de ruidos de baja frecuencia (por ejemplo, el zumbido de 60 Hz).

La frecuencia de corte para un filtro RC pasa altas es:
\begin{equation}
f_c = \frac{1}{2 \pi RC}
\end{equation}
donde, por debajo de $ f_c $, las señales se atenúan y, por encima de $f_c $, las señales pasan con mínima atenuación.

# Filtro Pasa Bandas

Un filtro pasa bandas permite el paso de un rango específico de frecuencias y atenúa (o reduce) las frecuencias fuera de este rango, tanto las más bajas como las más altas. Este tipo de filtro es útil en aplicaciones donde se necesita aislar una frecuencia o banda de frecuencias en particular, como en sistemas de audio, comunicación y en sintonización de señales de radio.

Ancho de Banda: El rango de frecuencias que el filtro permite pasar, calculado como:
\begin{equation}
    AB = f_{c2} - f_{c1}
\end{equation}

Este ancho de banda determina cuántas frecuencias pasan a través del filtro. Cuanto más cercano estén $ f_{c1} $ y $ f_{c2} $, más estrecho será el ancho de banda.



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, freqz, lfilter

# Parámetros generales
fs = 1000  # Frecuencia de muestreo en Hz

# Función para diseñar un filtro RC pasa bajas o pasa altas
def diseño_filtro_RC(tipo, fc, fs, orden=1):
    nyquist = fs / 2
    normal_fc = fc / nyquist
    b, a = butter(orden, normal_fc, btype=tipo, analog=False)
    return b, a

# Diseñamos el filtro pasa altas y pasa bajas para combinar en un pasa bandas
b_pasa_altas, a_pasa_altas = diseño_filtro_RC('high', 20, fs)  # Pasa altas con fc = 20 Hz
b_pasa_bajas, a_pasa_bajas = diseño_filtro_RC('low', 100, fs)  # Pasa bajas con fc = 100 Hz

# Respuesta en frecuencia para cada sección del filtro pasa bandas
w_pasa_altas, h_pasa_altas = freqz(b_pasa_altas, a_pasa_altas, worN=8000, fs=fs)
w_pasa_bajas, h_pasa_bajas = freqz(b_pasa_bajas, a_pasa_bajas, worN=8000, fs=fs)

# Filtro pasa bandas aplicando ambos filtros en secuencia
w_pasa_bandas = w_pasa_bajas  # La frecuencia será la misma para ambas secciones
h_pasa_bandas = h_pasa_altas * h_pasa_bajas  # Combinación de la respuesta en frecuencia

# Graficamos la respuesta en frecuencia de cada filtro y resaltamos la región de paso
plt.figure(figsize=(12, 8))

# Filtro pasa altas (20 Hz)
plt.subplot(3, 1, 1)
plt.plot(w_pasa_altas, abs(h_pasa_altas), label="Sección Pasa Altas (fc=20 Hz)", color='green')
plt.axvline(20, color='red', linestyle='--', label='Frecuencia de corte 20 Hz')
plt.fill_between(w_pasa_altas, abs(h_pasa_altas), where=(w_pasa_altas >= 20), color='green', alpha=0.3)
plt.xlabel("Frecuencia [Hz]")
plt.ylabel("Ganancia")
plt.title("Respuesta en frecuencia - Sección Pasa Altas")
plt.grid()
plt.legend()

# Filtro pasa bajas (100 Hz)
plt.subplot(3, 1, 2)
plt.plot(w_pasa_bajas, abs(h_pasa_bajas), label="Sección Pasa Bajas (fc=100 Hz)", color='orange')
plt.axvline(100, color='blue', linestyle='--', label='Frecuencia de corte 100 Hz')
plt.fill_between(w_pasa_bajas, abs(h_pasa_bajas), where=(w_pasa_bajas <= 100), color='orange', alpha=0.3)
plt.xlabel("Frecuencia [Hz]")
plt.ylabel("Ganancia")
plt.title("Respuesta en frecuencia - Sección Pasa Bajas")
plt.grid()
plt.legend()

# Filtro pasa bandas (combinación de pasa altas y pasa bajas)
plt.subplot(3, 1, 3)
plt.plot(w_pasa_bandas, abs(h_pasa_bandas), label="Filtro Pasa Bandas RC (20-100 Hz)", color='purple')
plt.axvline(20, color='red', linestyle='--', label='Frecuencia de corte 20 Hz')
plt.axvline(100, color='blue', linestyle='--', label='Frecuencia de corte 100 Hz')
plt.fill_between(w_pasa_bandas, abs(h_pasa_bandas), where=((w_pasa_bandas >= 20) & (w_pasa_bandas <= 100)), color='purple', alpha=0.3)
plt.xlabel("Frecuencia [Hz]")
plt.ylabel("Ganancia")
plt.title("Respuesta en frecuencia - Filtro Pasa Bandas RC")
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```

# Amplificador de Instrumentación

Un amplificador de instrumentación (INA) es un amplificador diferencial de alta precisión, diseñado para amplificar señales pequeñas en presencia de ruido o señales de modo común. Los INAs son ampliamente utilizados en aplicaciones de medición, donde se requiere amplificar solo la diferencia entre dos señales de entrada $ V_{in1} $ y $ V_{in2} $, mientras se rechaza el ruido común a ambas entradas.

La característica principal de un amplificador de instrumentación es su capacidad para rechazar señales de modo común gracias a su rechazo de modo común (CMRR).

# Ganancia del Amplificador de Instrumentación

La ganancia $ G $ de un amplificador de instrumentación se calcula mediante la siguiente fórmula:
\begin{equation}
G = 1 + \frac{50 \, \text{k}\Omega}{R_G}
\end{equation}

Al ajustar el valor de $ R_G $, se controla la ganancia del amplificador, lo cual permite adaptar el amplificador de instrumentación a diferentes aplicaciones.

# Voltaje de Salida

El voltaje de salida $ V_{out} $ del amplificador de instrumentación está dado por:
\begin{equation}
V_{out} = G \cdot (V_{in1} - V_{in2})
\end{equation}

La salida $ V_{out} $ es una señal amplificada de la diferencia de los voltajes de entrada, permitiendo así medir señales pequeñas en presencia de ruido.


```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de entrada
V_in1 = 1.0  # Voltaje en la entrada + en voltios
V_in2 = 0.5  # Voltaje en la entrada - en voltios
Vin_diff = V_in1 - V_in2  # Voltaje diferencial de entrada

# Valores de resistencia R_G para explorar distintas ganancias
Rg_values = np.array([100, 500, 1000, 5000, 10000])  # Valores en ohmios

# Creamos listas para almacenar los resultados
gain_values = []
Vout_values = []

# Cálculos para cada valor de R_G
for Rg in Rg_values:
    # Calculamos la ganancia
    G = 1 + (50000 / Rg)
    # Calculamos el voltaje de salida
    V_out = G * Vin_diff

    # Guardamos los resultados en las listas
    gain_values.append(G)
    Vout_values.append(V_out)

# Graficar la ganancia en función de R_G
plt.figure(figsize=(12, 6))

# Gráfico de ganancia en función de R_G
plt.subplot(2, 1, 1)
plt.plot(Rg_values, gain_values, marker='o', color='blue', label="Ganancia (G)")
plt.xlabel("Resistencia $R_G$ (ohmios)")
plt.ylabel("Ganancia (G)")
plt.title("Ganancia del Amplificador de Instrumentación en función de $R_G$")
plt.legend()
plt.grid(True)

# Gráfico de voltaje de salida en función de R_G
plt.subplot(2, 1, 2)
plt.plot(Rg_values, Vout_values, marker='o', color='red', label="Voltaje de Salida $V_{out}$")
plt.xlabel("Resistencia $R_G$ (ohmios)")
plt.ylabel("Voltaje de Salida $V_{out}$ (V)")
plt.title("Voltaje de Salida del Amplificador de Instrumentación en función de $R_G$")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# Mostramos los resultados en consola
for i, Rg in enumerate(Rg_values):
    print(f"R_G = {Rg} ohmios: Ganancia = {gain_values[i]:.2f}, V_out = {Vout_values[i]:.2f} V")
```

**Integral Definida**

La integral definida de una función $ f(x) $ en el intervalo $[a, b]$ representa el área bajo la curva de $ f(x) $ entre los puntos $ x = a$ y $ x = b $. Se denota como:
\begin{equation}
\int_{a}^{b} f(x) \, dx
\end{equation}

El valor de la integral definida nos proporciona un número que representa el área acumulada bajo $ f(x) $ entre $ x = a $ y $ x = b $.

**Integral Indefinida**

La \textbf{integral indefinida} de una función $ f(x) $ representa el conjunto de todas las antiderivadas de $ f(x) $. Es decir, la integral indefinida es una familia de funciones $ F(x) + C $ tal que la derivada de $ F(x) $ es $ f(x) $. Se denota como:
\begin{equation}
\int f(x) \, dx = F(x) + C
\end{equation}

**Integral Convergente (No Divergente)**

Una integral convergente es aquella cuyo valor se aproxima a un número finito. Esto es particularmente importante en las integrales impropias, donde el intervalo de integración es infinito o la función tiene puntos de discontinuidad en el intervalo de integración.


**Integral Divergente**

Una integral divergente es aquella cuyo valor no converge a un número finito. Las integrales pueden divergir en el caso de integrales impropias o cuando el intervalo de integración incluye puntos de discontinuidad.



# Integrales Definidas

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
from sympy import symbols, integrate, sin, cos, exp

# Definimos la función a integrar (ejemplo: f(x) = sin(x))
def f(x):
    return np.sin(x)

# Intervalo de integración para la integral definida
a, b = 0, np.pi  # Integrar de 0 a π

# Cálculo de la integral definida con scipy
integral_definida, error = quad(f, a, b)
print(f"Valor de la integral definida de sin(x) desde 0 a π: {integral_definida:.4f}")

# Visualización de la integral definida (área bajo la curva)
x = np.linspace(-1, 4, 100)
y = f(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label=r"$f(x) = \sin(x)$", color='blue')
plt.fill_between(x, y, where=[(x_i >= a) & (x_i <= b) for x_i in x], color='cyan', alpha=0.3, label="Área bajo la curva")
plt.axhline(0, color='black', linewidth=0.5)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Integral Definida de sin(x) en el intervalo [0, π]")
plt.legend()
plt.grid()
plt.show()

# Cálculo de la integral indefinida usando sympy
x_sym = symbols('x')
f_sym = sin(x_sym)
integral_indefinida = integrate(f_sym, x_sym)
print(f"Integral indefinida de sin(x): ∫sin(x) dx = {integral_indefinida} + C")
```

# Integrales Indefinidas

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, integrate, sin, cos, exp, latex, lambdify

# Definimos la variable simbólica
x = symbols('x')

# Lista de funciones a integrar
funciones = [
    x**2,           # f(x) = x^2
    sin(x),         # f(x) = sin(x)
    exp(x),         # f(x) = e^x
    cos(x) + x**2   # f(x) = cos(x) + x^2
]

# Diccionario para almacenar las integrales
integrales = {}

# Calcular las integrales indefinidas
for f in funciones:
    integral_f = integrate(f, x)
    integrales[f] = integral_f
    print(f"Integral indefinida de {f}: ∫{latex(f)} \, dx = {latex(integral_f)} + C")

# Graficar las funciones y sus antiderivadas
x_vals = np.linspace(-5, 5, 400)  # Valores de x para la gráfica

plt.figure(figsize=(12, len(funciones) * 4))

for i, f in enumerate(funciones, 1):
    # Convertir la función y la integral a funciones numéricas usando lambdify
    f_num = lambdify(x, f, 'numpy')
    F_num = lambdify(x, integrales[f], 'numpy')

    # Graficar la función original
    plt.subplot(len(funciones), 2, 2 * i - 1)
    plt.plot(x_vals, f_num(x_vals), label=f'$f(x) = {latex(f)}$', color='blue')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title(f"Función Original: $f(x) = {latex(f)}$")
    plt.legend()
    plt.grid(True)

    # Graficar la integral indefinida
    plt.subplot(len(funciones), 2, 2 * i)
    plt.plot(x_vals, F_num(x_vals), label=f'$\int {latex(f)} \, dx = {latex(integrales[f])} + C$', color='red')
    plt.xlabel('x')
    plt.ylabel('F(x)')
    plt.title(f"Integral Indefinida: $\int {latex(f)} \, dx$")
    plt.legend()
    plt.grid(True)

plt.tight_layout()
plt.show()
```

# Integrales que Divergen o No Divergen

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Definir la función a integrar
def f(x):
    return np.exp(-x**2)  # Función gaussiana

# Calcular la integral desde -∞ hasta ∞
integral, error = quad(f, -np.inf, np.inf)

# Verificar si la integral converge
if np.isfinite(integral) and error < 1e-4:
    print(f"La integral converge y su valor es: {integral:.4f}")
else:
    print("La integral diverge.")

# Graficar la función
x_vals = np.linspace(-5, 5, 400)
y_vals = f(x_vals)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label=r"$f(x) = e^{-x^2}$", color='blue')
plt.fill_between(x_vals, y_vals, alpha=0.2, color='cyan', label="Área bajo la curva")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Integral de $f(x) = e^{-x^2}$ en el intervalo $(-\infty, \infty)$")
plt.legend()
plt.grid(True)
plt.show()
```

# Máximos y Mínimos

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, diff, solve, lambdify, sympify, log, exp

# Solicitar al usuario que ingrese la función
funcion_str = input("Introduce la función en términos de x (por ejemplo, 'x**3 - 3*x**2 + 4' o 'exp(x) - x*log(x)'): ")
x = symbols('x')
f = sympify(funcion_str)  # Convertir la cadena de texto en una expresión simbólica

# Calcular la primera y segunda derivada
f_prime = diff(f, x)
f_double_prime = diff(f_prime, x)

# Encontrar los puntos críticos resolviendo f'(x) = 0
puntos_criticos = solve(f_prime, x)

# Determinar si cada punto crítico es un máximo, mínimo o punto de inflexión
extremos = []
for punto in puntos_criticos:
    valor_segunda_derivada = f_double_prime.subs(x, punto)
    if valor_segunda_derivada > 0:
        extremos.append((punto, "Mínimo"))
    elif valor_segunda_derivada < 0:
        extremos.append((punto, "Máximo"))
    else:
        extremos.append((punto, "Inconcluso"))

# Mostrar los puntos críticos y su clasificación
for punto, tipo in extremos:
    print(f"Punto crítico en x = {punto}: {tipo}")

# Convertir la función a una función numérica para graficar
f_num = lambdify(x, f, 'numpy')

# Definir el rango de valores de x para la gráfica (evitar logaritmos de valores no permitidos)
x_vals = np.linspace(-10, 10, 400)
x_vals = x_vals[x_vals > 0.1] if 'log' in funcion_str else x_vals  # Evita valores negativos en logaritmos

# Graficar la función y marcar los extremos
y_vals = f_num(x_vals)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label=f'$f(x) = {funcion_str}$', color='blue')

# Marcar los máximos y mínimos en la gráfica
for punto, tipo in extremos:
    if punto.is_real:
        y_punto = f_num(float(punto))
        color = 'red' if tipo == "Máximo" else 'green'
        plt.plot(float(punto), y_punto, 'o', color=color, label=f'{tipo} en x = {punto}')

# Configurar la gráfica
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title(f"Máximos y Mínimos de la Función $f(x) = {funcion_str}$")
plt.legend()
plt.grid(True)
plt.show()
```

# Gráficos 3D

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, sympify, lambdify, log, exp
from mpl_toolkits.mplot3d import Axes3D

# Solicitar la función al usuario
funcion_str = input("Introduce la función en términos de x e y (por ejemplo, 'x**2 + y**2 - 4*x - 2*y' o 'exp(x) - y*log(y)'): ")

# Definir las variables simbólicas
x, y = symbols('x y')
# Convertir la cadena de texto en una expresión simbólica
f = sympify(funcion_str)

# Convertir la función simbólica a una función numérica para evaluarla en numpy
f_num = lambdify((x, y), f, 'numpy')

# Rango de valores para x e y
x_vals = np.linspace(-5, 5, 100)
y_vals = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x_vals, y_vals)

# Calcular Z usando la función numérica
# Agregar un manejo de errores para evitar valores no definidos (por ejemplo, log(0) o log(negativos))
try:
    Z = f_num(X, Y)
except ValueError as e:
    print(f"Error al evaluar la función: {e}")
    Z = np.zeros_like(X)  # En caso de error, usar una superficie plana

# Graficar la función en 3D
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.7)

# Configurar la gráfica
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("f(x, y)")
plt.title(f"Gráfica de la Función $f(x, y) = {funcion_str}$")
plt.show()
```

# Respuesta de Frecuencias de Filtros

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, cheby1, cheby2, ellip, bessel, freqz

# Definir el diseño del filtro
def diseño_filtro(tipo_filtro, fc, fs, orden=2, ripple_db=1, atenuacion_db=40):
    nyquist = 0.5 * fs  # Frecuencia de Nyquist
    normal_fc = fc / nyquist

    if tipo_filtro == 'butterworth':
        b, a = butter(orden, normal_fc, btype='low', analog=False)
    elif tipo_filtro == 'chebyshev1':
        b, a = cheby1(orden, ripple_db, normal_fc, btype='low', analog=False)
    elif tipo_filtro == 'chebyshev2':
        b, a = cheby2(orden, atenuacion_db, normal_fc, btype='low', analog=False)
    elif tipo_filtro == 'elliptic':
        b, a = ellip(orden, ripple_db, atenuacion_db, normal_fc, btype='low', analog=False)
    elif tipo_filtro == 'bessel':
        b, a = bessel(orden, normal_fc, btype='low', analog=False, norm='phase')
    else:
        raise ValueError("Tipo de filtro no soportado.")
    return b, a

# Definir las características del filtro
tipo_filtro = input("Elige el tipo de filtro ('butterworth', 'chebyshev1', 'chebyshev2', 'elliptic', 'bessel'): ")
fc = float(input("Introduce la frecuencia de corte (Hz): "))
fs = float(input("Introduce la frecuencia de muestreo (Hz): "))
orden = int(input("Introduce el orden del filtro (1 o 2): "))

# Parámetros adicionales para Chebyshev y Elíptico
ripple_db = 1  # en dB (sólo para Chebyshev tipo 1 y Elíptico)
atenuacion_db = 40  # en dB (sólo para Chebyshev tipo 2 y Elíptico)

# Calcular los coeficientes del filtro
try:
    b, a = diseño_filtro(tipo_filtro, fc, fs, orden, ripple_db, atenuacion_db)
except ValueError as e:
    print(f"Error: {e}")
else:
    # Calcular la respuesta en frecuencia del filtro
    w, h = freqz(b, a, worN=8000, fs=fs)

    # Graficar la respuesta en frecuencia
    plt.figure(figsize=(12, 6))
    plt.plot(w, 20 * np.log10(abs(h)), 'b')
    plt.title(f"Respuesta en Frecuencia del Filtro {tipo_filtro.capitalize()} - Orden {orden}")
    plt.xlabel("Frecuencia (Hz)")
    plt.ylabel("Ganancia (dB)")
    plt.grid()
    plt.axvline(fc, color='red', linestyle='--', label=f'Frecuencia de Corte: {fc} Hz')
    plt.legend()
    plt.show()
```
