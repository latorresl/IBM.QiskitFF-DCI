##Matriz inversa
La matriz inversa tiene muchas propiedades interesantes, como:


*  Multiplicar la matriz inversa por la matriz original resulta en la matriz identidad: \\
$B^{-1} \times B = \mathbb{I}$ \\
  donde $\mathbb{I}$ es una matriz slo con unos en su diagonal:
  \begin{equation}
  \mathbb{I}=
  \begin{pmatrix}
  1 & 0 \\
  0 & 1 \\
  \end{pmatrix}
   \space  \space \space  
  \mathbb{I}=
  \begin{pmatrix}
  1 & 0 & 0 \\
  0 & 1 & 0 \\
  0 & 0 & 1
  \end{pmatrix}
  \end{equation}
*  Encontrar la matriz inversa nos permite resolver un sistema de ecuaciones como el siguiente

  $a_1 x + a_2 y + a_3 z=c$ \\
  $b_1 x+ b_2 y + b_3z=d$ \\
  $c_1 x + c_2 y + c_3 z =e$ \\

  Podemos expresarlo en un sistema matricial:
\begin{equation}
  \begin{pmatrix}
  a_1 & a_2 & a_3 \\
  b_1 & b_2 & b_3 \\
  c_1 & c_2 & c_3 \\
  \end{pmatrix}
  \times
  \begin{pmatrix}
  x \\ y \\ z
  \end{pmatrix}
  =
  \begin{pmatrix}
  c \\ d \\ e
  \end{pmatrix}
\end{equation}
  Esto es:
  \begin{equation}
  A \times \vec{x} = \vec{c}
  \end{equation}
  Si encontramos la matriz inversa, $A^{-1}$, y calculamos el producto matricial con $\vec{c}$ (el trmino de la derecha, llamado *vector de coeficientes*) el resultado ser una matriz cuyas componentes son iguales a los valores que toman $x$, $y$ y $z$. \\
  Esto es, multiplicando por $A^{-1}$ la ecuacin anterior:
  \begin{equation}
  A^{-1} \times A \times \vec{x} = A^{-1} \times \vec{c}
  \end{equation}

  \begin{equation}
  \mathbb{I} \times \vec{x} = A^{-1} \times \vec{c}
  \end{equation}
  Luego, la matriz identidad multiplicada por el vector $\vec{x}$ resulta en el mismo vector, y obtenemos:
  \begin{equation}
  \vec{x} = A^{-1} \times \vec{c}
  \end{equation}






###**Podemos aplicar estos conceptos y resolver fcilmente un sistema de ecuaciones con el uso de la librera Numpy**

Queremos resolver el siguiente sistema de ecuaciones: \\
\begin{equation}
x+y+z=4\\
x+2y+4z=0 \\
x+3y+9z=12
\end{equation}
ste lo podemos representar como:
\begin{equation}
  \begin{pmatrix}
  1 & 1 & 1 \\
  1 & 2 & 4 \\
  1 & 3 & 9 \\
  \end{pmatrix}
  \times
  \begin{pmatrix}
  x\\ y \\ z
  \end{pmatrix}
  =
  \begin{pmatrix}
  4\\ 0 \\ 12
  \end{pmatrix}
\end{equation}

```python
A=np.array([[1,1,1],[1,2,4],[1,3,9]])
print("A=")
print(A)
c=np.array([[4],[0],[12]]) #Vector de coeficientes
x=linalg.solve(A,c) #Vector x, valores de las incgnitas
print("x=",x[0]," y=",x[1]," z=",x[2])
```

##**Teorema de la descomposicin espectral**

Si A es una matriz cuadrada (de dimensiones n $\times$ n), y simtrica, se tienen los siguientes resultados:


1.   Los valores propios (eigenvalores o autovalores) son reales
2.   Los vectores propios (eigenvectores o autovectores) son ortogonales entre s
3.   La matriz original A se puede diagonalizar, es decir, existe una matriz P que cumple que:
  $P^{-1} \times A \times P = D$ \\
  donde D es una matriz diagonal.

Ejemplos de matriz diagonal:
\begin{equation}
\begin{pmatrix}
i & 0 & 0 \\
0 & j & 0 \\
0 & 0 & k
\end{pmatrix}
\end{equation}
Finalmente, todo ello para *descomponer* la matriz A en su *espectro* (otras tantas matrices) que se forman con los valores y vectores propios de A.

```python
A=np.array([[7,-2],[-2,4]])
print("A=")
print(A)
eigenvalores,eigenvectores=linalg.eig(A)
# Separamos primero los eigenvalores
a=eigenvalores[0]
b=eigenvalores[1]
# Separamos los eigenvectores
x=eigenvectores[:,0]
y=eigenvectores[:,1]

print("eigenavalor 1 = ", a, ", eigenvalor 2 = ", b)
print("eigenvector 1=",x)
print("eigenvector 2=", y)

print("Producto punto =",np.vdot(x,y)) #Estos vectores son ortogonales entre s, pues su producto punto es cero
```

# Oscilador Armnico Amortiguado (OAA)

El oscilador armnico amortiguado es un modelo matemtico para describir el movimiento de objetos que oscilan respecto a un punto en el espacio y que eventualmente se detienen. El parmetro de amortiguamiento $\beta=b/2m$ en la ecuacin
\begin{equation}
m\frac{d^2x}{dt^2}+b\frac{dx}{dt}+kx=0
\end{equation}
es caracterstico de cada movimiento de este tipo. Se obtiene
\begin{equation}
\frac{d^2x}{dt^2}+2\beta\frac{dx}{dt}+w_o^2x=0
\end{equation}
donde $w_o^2=k/m$ es la frecuencia natural de oscilacin del objeto.

La solucin general de esta ecuacin diferencial de movimiento es
\begin{equation}
x(t)=e^{-\beta t}\left[ A\text{exp}\left(t\sqrt{\beta^2-w_o^2}\right) + B\text{exp}\left(-t\sqrt{\beta^2-w_o^2}\right)  \right]
\end{equation}

Y segn la relacin entre el parmetro de amortiguamiento $\beta$ y la frecuencia natural de oscilacin $w_o$ esta solucin, y el fenmeno fsico, caen en determinado caso. Analizaremos el caso de oscilaciones subamortiguadas donde $w_o^2>\beta^2$.

La solucin, que es el modelo no lineal que ajustaremos a los datos, es
\begin{equation}
x(t)=Ae^{-\beta t} \cos{(wt-\delta)}
\end{equation}

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.optimize import curve_fit
```

```python
dataset = pd.read_csv('https://raw.githubusercontent.com/latorresl/PhyLab-Statistics/refs/heads/main/The%20Data%20Sets/OAAm.csv')
x = dataset['tOAAm'] # Posicin en x
y = dataset['xOAAm'] # Posicin en y
```

```python
# Definimos un modelo. Es una funcin, cuyo primer argumento son los valores
# del eje x y el resto de los argumentos son los parmetros a hallar

def OAA(x, a, b, c, d):
    return a * np.exp(-b*x) * np.cos(c * x + d)

# Parmetros iniciales con los que vamos a iniciar el proceso de ajuste (meras aproximaciones)
parametros_iniciales=[3, 0.25, 3, -np.pi]

# Hacemos el ajuste con curve_fit
popt, pcov = curve_fit(OAA, x, y, p0=parametros_iniciales)
incertidumbre = np.sqrt(np.diag(pcov))

a,b,c,d = popt

# curve_fit devuelve dos resultados. El primero (popt) son los
# parmetros ptimos hallados. El segundo (pcov) es la matriz de
# covarianza de los parametros hallados.

# Se genera el intervalo de graficado
x_modelo  = np.linspace(min(x), max(x), 1000)

# Se definen las funciones contenedoras
cont1 = a*np.exp(-b*x)
cont2 = -a*np.exp(-b*x)

# popt tiene los parmetros hallados en orden
print('Parmetros ordenados: {}'.format(popt))

# De la matriz de covarianza podemos obtener los valores de desviacion
# estandar de los parametros ubicados en su diagonal
pstd = np.sqrt(np.diag(pcov))

nombres_de_param=['a','b','c','d']
print('\nParmetros hallados:')
for i,param in enumerate(popt):
    print('{:s} = {}  {}'.format( nombres_de_param[i] , param , pstd[i]/2) )

print('\nCoeficientes: {}\nIncertidumbre: {}\n\nMatriz de covarianza:\n {}'.format(popt,incertidumbre,pcov))

# Calcular R^2
y_mean = np.mean(y)
TSS = np.sum((y - y_mean)**2)
y_pred = OAA(x, a, b, c, d)
RSS = np.sum((y - y_pred)**2)
r_squared = 1 - (RSS / TSS)

print(f'\nCoeficiente R^2: {r_squared}')
```

### Covarianza y matriz de covarianza
La covarianza es una medida de cunto varan de manera conjunta dos variables aleatorias.

\begin{equation}
\operatorname{Cov}(X_i, X_j) = \mathbb{E}[(X_i - \mathbb{E}[X_i])(X_j - \mathbb{E}[X_j])]
\end{equation}

De modo que la matriz de covarianza es

\begin{equation}
\Sigma = \begin{pmatrix}
\operatorname{Cov}(X_1, X_1) & \operatorname{Cov}(X_1, X_2) & \cdots & \operatorname{Cov}(X_1, X_n) \\
\operatorname{Cov}(X_2, X_1) & \operatorname{Cov}(X_2, X_2) & \cdots & \operatorname{Cov}(X_2, X_n) \\
\vdots & \vdots & \ddots & \vdots \\
\operatorname{Cov}(X_n, X_1) & \operatorname{Cov}(X_n, X_2) & \cdots & \operatorname{Cov}(X_n, X_n)
\end{pmatrix}
\end{equation}

La covarianza de dos variables aleatorias iguales es la varianza de esa variable, $\operatorname{Cov}(X_i, X_j)=\operatorname{Var}(X)=\sigma_X^2$.

# Catstrofe Ultravioleta



Cuando la teora de la electrodinmica clsica fall al explicar la emisin electromagntica de un cuerpo en equilibrio trmico con el ambiente con la ecuacin de Rayleigh-Jeans
\begin{equation}
I()=\frac{8}{c^3}kT^2
\end{equation}
que fallaba al describir los resultados experimentales sobre la intensidad de la radiacin para una frecuencia $$, Max Planck propuso que esta intensidad estaba descrita por la ecuacin
\begin{equation}
I()=\frac{8h}{c^3}\frac{^3}{e^{h/kT}-1}
\end{equation}
Donde introdujo el concepto de cuanto. Si la energa de la radiacin en la ecuacin anterior es
\begin{equation}
=\frac{h}{e^{h/kT}-1}
\end{equation}
el cuanto de energa para una radiacin de frecuencia $$ es $=h$. Graficaremos el comportamiento de ambos modelos, notando que el modelo de Planck se ajust a los resultados de los experimentos realizados en esa poca.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
h = 6.626e-34  # Constante de Planck (Js)
c = 3.0e8      # Velocidad de la luz (m/s)
k = 1.38e-23   # Constante de Boltzmann (J/K)
pi = np.pi

T = 1500

# Rango de frecuencias
f = np.linspace(1e10, 1e15, 1000)
```

```python
# Ecuacin de Rayleigh-Jeans
def rayleigh_jeans(f, T):
    return (8 * pi * k * T * f**2) / (c**3)

# Ecuacin de Planck
def planck(f, T):
    return (8 * pi * h * f**3) / (c**3 * (np.exp(h * f / (k * T)) - 1))

# Calcular las densidades de energa espectral
G_f_rayleigh = rayleigh_jeans(f, T)
u_f_planck = planck(f, T)
```

```python
plt.rcParams['font.family'] = 'serif'
font = {'family': 'serif', 'color': 'black', 'weight': 'light', 'size': 14}

plt.figure(figsize=(8, 6))
plt.plot(f, G_f_rayleigh, label="Rayleigh-Jeans", color='deeppink', linestyle='--', linewidth=4)
plt.plot(f, u_f_planck, label="Planck", color='deeppink', linewidth=4)
plt.xlabel("Frequency, v (Hz)")
plt.ylabel("Spectral energy density, u(v)dv")
plt.xlim(0, 4e14)  # Limitar el rango de frecuencias
plt.ylim(0, max(u_f_planck) * 2)  # Limitar el rango de la densidad de energa
plt.grid(True)

# Ajuste de escala en el eje x para mltiplos de 10^14
plt.gca().set_yticklabels([])
xticks = np.arange(0, 5e14, 1e14)  # Marcas en mltiplos de 10^14
plt.xticks(xticks)  # Aplicar marcas en el eje x
plt.gca().ticklabel_format(style='sci', axis='x', scilimits=(14, 14))

plt.gca().set_facecolor('pink')
plt.grid(color='white', linewidth=1.5, linestyle='-')

y_center = max(u_f_planck)
plt.text(1.1e14, y_center * 1.5, 'Rayleigh-Jeans', fontsize=15, ha='center')
plt.text(3e14, y_center * 0.2, 'M. Planck', fontsize=15, ha='center')
plt.show()
```

# Filtro Pasa Bajas

Un filtro pasa bajas permite el paso de frecuencias bajas y atena (reduce) las frecuencias altas a partir de una frecuencia de corte especfica ($f_c$). Es comn en aplicaciones donde solo se requieren frecuencias bajas, como en la eliminacin de ruido de alta frecuencia.

La frecuencia de corte se calcula como:
\begin{equation}
f_c = \frac{1}{2 \pi RC}
\end{equation}
donde R es la resistencia y C es la capacitancia. Por debajo de $f_c$, la seal pasa con mnima atenuacin, y por encima de $f_c$, se atena cada vez ms.

# Filtro Pasa Altas

Un filtro pasa altas permite el paso de frecuencias altas y atena las frecuencias bajas. Es til en aplicaciones donde solo se quieren frecuencias altas, como en la eliminacin de ruidos de baja frecuencia (por ejemplo, el zumbido de 60 Hz).

La frecuencia de corte para un filtro RC pasa altas es:
\begin{equation}
f_c = \frac{1}{2 \pi RC}
\end{equation}
donde, por debajo de $ f_c $, las seales se atenan y, por encima de $f_c $, las seales pasan con mnima atenuacin.

# Filtro Pasa Bandas

Un filtro pasa bandas permite el paso de un rango especfico de frecuencias y atena (o reduce) las frecuencias fuera de este rango, tanto las ms bajas como las ms altas. Este tipo de filtro es til en aplicaciones donde se necesita aislar una frecuencia o banda de frecuencias en particular, como en sistemas de audio, comunicacin y en sintonizacin de seales de radio.

Ancho de Banda: El rango de frecuencias que el filtro permite pasar, calculado como:
\begin{equation}
    AB = f_{c2} - f_{c1}
\end{equation}

Este ancho de banda determina cuntas frecuencias pasan a travs del filtro. Cuanto ms cercano estn $ f_{c1} $ y $ f_{c2} $, ms estrecho ser el ancho de banda.



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, freqz, lfilter

# Parmetros generales
fs = 1000  # Frecuencia de muestreo en Hz

# Funcin para disear un filtro RC pasa bajas o pasa altas
def diseo_filtro_RC(tipo, fc, fs, orden=1):
    nyquist = fs / 2
    normal_fc = fc / nyquist
    b, a = butter(orden, normal_fc, btype=tipo, analog=False)
    return b, a

# Diseamos el filtro pasa altas y pasa bajas para combinar en un pasa bandas
b_pasa_altas, a_pasa_altas = diseo_filtro_RC('high', 20, fs)  # Pasa altas con fc = 20 Hz
b_pasa_bajas, a_pasa_bajas = diseo_filtro_RC('low', 100, fs)  # Pasa bajas con fc = 100 Hz

# Respuesta en frecuencia para cada seccin del filtro pasa bandas
w_pasa_altas, h_pasa_altas = freqz(b_pasa_altas, a_pasa_altas, worN=8000, fs=fs)
w_pasa_bajas, h_pasa_bajas = freqz(b_pasa_bajas, a_pasa_bajas, worN=8000, fs=fs)

# Filtro pasa bandas aplicando ambos filtros en secuencia
w_pasa_bandas = w_pasa_bajas  # La frecuencia ser la misma para ambas secciones
h_pasa_bandas = h_pasa_altas * h_pasa_bajas  # Combinacin de la respuesta en frecuencia

# Graficamos la respuesta en frecuencia de cada filtro y resaltamos la regin de paso
plt.figure(figsize=(12, 8))

# Filtro pasa altas (20 Hz)
plt.subplot(3, 1, 1)
plt.plot(w_pasa_altas, abs(h_pasa_altas), label="Seccin Pasa Altas (fc=20 Hz)", color='green')
plt.axvline(20, color='red', linestyle='--', label='Frecuencia de corte 20 Hz')
plt.fill_between(w_pasa_altas, abs(h_pasa_altas), where=(w_pasa_altas >= 20), color='green', alpha=0.3)
plt.xlabel("Frecuencia [Hz]")
plt.ylabel("Ganancia")
plt.title("Respuesta en frecuencia - Seccin Pasa Altas")
plt.grid()
plt.legend()

# Filtro pasa bajas (100 Hz)
plt.subplot(3, 1, 2)
plt.plot(w_pasa_bajas, abs(h_pasa_bajas), label="Seccin Pasa Bajas (fc=100 Hz)", color='orange')
plt.axvline(100, color='blue', linestyle='--', label='Frecuencia de corte 100 Hz')
plt.fill_between(w_pasa_bajas, abs(h_pasa_bajas), where=(w_pasa_bajas <= 100), color='orange', alpha=0.3)
plt.xlabel("Frecuencia [Hz]")
plt.ylabel("Ganancia")
plt.title("Respuesta en frecuencia - Seccin Pasa Bajas")
plt.grid()
plt.legend()

# Filtro pasa bandas (combinacin de pasa altas y pasa bajas)
plt.subplot(3, 1, 3)
plt.plot(w_pasa_bandas, abs(h_pasa_bandas), label="Filtro Pasa Bandas RC (20-100 Hz)", color='purple')
plt.axvline(20, color='red', linestyle='--', label='Frecuencia de corte 20 Hz')
plt.axvline(100, color='blue', linestyle='--', label='Frecuencia de corte 100 Hz')
plt.fill_between(w_pasa_bandas, abs(h_pasa_bandas), where=((w_pasa_bandas >= 20) & (w_pasa_bandas <= 100)), color='purple', alpha=0.3)
plt.xlabel("Frecuencia [Hz]")
plt.ylabel("Ganancia")
plt.title("Respuesta en frecuencia - Filtro Pasa Bandas RC")
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```

# Amplificador de Instrumentacin

Un amplificador de instrumentacin (INA) es un amplificador diferencial de alta precisin, diseado para amplificar seales pequeas en presencia de ruido o seales de modo comn. Los INAs son ampliamente utilizados en aplicaciones de medicin, donde se requiere amplificar solo la diferencia entre dos seales de entrada $ V_{in1} $ y $ V_{in2} $, mientras se rechaza el ruido comn a ambas entradas.

La caracterstica principal de un amplificador de instrumentacin es su capacidad para rechazar seales de modo comn gracias a su rechazo de modo comn (CMRR).

# Ganancia del Amplificador de Instrumentacin

La ganancia $ G $ de un amplificador de instrumentacin se calcula mediante la siguiente frmula:
\begin{equation}
G = 1 + \frac{50 \, \text{k}\Omega}{R_G}
\end{equation}

Al ajustar el valor de $ R_G $, se controla la ganancia del amplificador, lo cual permite adaptar el amplificador de instrumentacin a diferentes aplicaciones.

# Voltaje de Salida

El voltaje de salida $ V_{out} $ del amplificador de instrumentacin est dado por:
\begin{equation}
V_{out} = G \cdot (V_{in1} - V_{in2})
\end{equation}

La salida $ V_{out} $ es una seal amplificada de la diferencia de los voltajes de entrada, permitiendo as medir seales pequeas en presencia de ruido.


```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros de entrada
V_in1 = 1.0  # Voltaje en la entrada + en voltios
V_in2 = 0.5  # Voltaje en la entrada - en voltios
Vin_diff = V_in1 - V_in2  # Voltaje diferencial de entrada

# Valores de resistencia R_G para explorar distintas ganancias
Rg_values = np.array([100, 500, 1000, 5000, 10000])  # Valores en ohmios

# Creamos listas para almacenar los resultados
gain_values = []
Vout_values = []

# Clculos para cada valor de R_G
for Rg in Rg_values:
    # Calculamos la ganancia
    G = 1 + (50000 / Rg)
    # Calculamos el voltaje de salida
    V_out = G * Vin_diff

    # Guardamos los resultados en las listas
    gain_values.append(G)
    Vout_values.append(V_out)

# Graficar la ganancia en funcin de R_G
plt.figure(figsize=(12, 6))

# Grfico de ganancia en funcin de R_G
plt.subplot(2, 1, 1)
plt.plot(Rg_values, gain_values, marker='o', color='blue', label="Ganancia (G)")
plt.xlabel("Resistencia $R_G$ (ohmios)")
plt.ylabel("Ganancia (G)")
plt.title("Ganancia del Amplificador de Instrumentacin en funcin de $R_G$")
plt.legend()
plt.grid(True)

# Grfico de voltaje de salida en funcin de R_G
plt.subplot(2, 1, 2)
plt.plot(Rg_values, Vout_values, marker='o', color='red', label="Voltaje de Salida $V_{out}$")
plt.xlabel("Resistencia $R_G$ (ohmios)")
plt.ylabel("Voltaje de Salida $V_{out}$ (V)")
plt.title("Voltaje de Salida del Amplificador de Instrumentacin en funcin de $R_G$")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# Mostramos los resultados en consola
for i, Rg in enumerate(Rg_values):
    print(f"R_G = {Rg} ohmios: Ganancia = {gain_values[i]:.2f}, V_out = {Vout_values[i]:.2f} V")
```

**Integral Definida**

La integral definida de una funcin $ f(x) $ en el intervalo $[a, b]$ representa el rea bajo la curva de $ f(x) $ entre los puntos $ x = a$ y $ x = b $. Se denota como:
\begin{equation}
\int_{a}^{b} f(x) \, dx
\end{equation}

El valor de la integral definida nos proporciona un nmero que representa el rea acumulada bajo $ f(x) $ entre $ x = a $ y $ x = b $.

**Integral Indefinida**

La \textbf{integral indefinida} de una funcin $ f(x) $ representa el conjunto de todas las antiderivadas de $ f(x) $. Es decir, la integral indefinida es una familia de funciones $ F(x) + C $ tal que la derivada de $ F(x) $ es $ f(x) $. Se denota como:
\begin{equation}
\int f(x) \, dx = F(x) + C
\end{equation}

**Integral Convergente (No Divergente)**

Una integral convergente es aquella cuyo valor se aproxima a un nmero finito. Esto es particularmente importante en las integrales impropias, donde el intervalo de integracin es infinito o la funcin tiene puntos de discontinuidad en el intervalo de integracin.


**Integral Divergente**

Una integral divergente es aquella cuyo valor no converge a un nmero finito. Las integrales pueden divergir en el caso de integrales impropias o cuando el intervalo de integracin incluye puntos de discontinuidad.



# Integrales Definidas

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
from sympy import symbols, integrate, sin, cos, exp

# Definimos la funcin a integrar (ejemplo: f(x) = sin(x))
def f(x):
    return np.sin(x)

# Intervalo de integracin para la integral definida
a, b = 0, np.pi  # Integrar de 0 a 

# Clculo de la integral definida con scipy
integral_definida, error = quad(f, a, b)
print(f"Valor de la integral definida de sin(x) desde 0 a : {integral_definida:.4f}")

# Visualizacin de la integral definida (rea bajo la curva)
x = np.linspace(-1, 4, 100)
y = f(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label=r"$f(x) = \sin(x)$", color='blue')
plt.fill_between(x, y, where=[(x_i >= a) & (x_i <= b) for x_i in x], color='cyan', alpha=0.3, label="rea bajo la curva")
plt.axhline(0, color='black', linewidth=0.5)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Integral Definida de sin(x) en el intervalo [0, ]")
plt.legend()
plt.grid()
plt.show()

# Clculo de la integral indefinida usando sympy
x_sym = symbols('x')
f_sym = sin(x_sym)
integral_indefinida = integrate(f_sym, x_sym)
print(f"Integral indefinida de sin(x): sin(x) dx = {integral_indefinida} + C")
```

# Integrales Indefinidas

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, integrate, sin, cos, exp, latex, lambdify

# Definimos la variable simblica
x = symbols('x')

# Lista de funciones a integrar
funciones = [
    x**2,           # f(x) = x^2
    sin(x),         # f(x) = sin(x)
    exp(x),         # f(x) = e^x
    cos(x) + x**2   # f(x) = cos(x) + x^2
]

# Diccionario para almacenar las integrales
integrales = {}

# Calcular las integrales indefinidas
for f in funciones:
    integral_f = integrate(f, x)
    integrales[f] = integral_f
    print(f"Integral indefinida de {f}: {latex(f)} \, dx = {latex(integral_f)} + C")

# Graficar las funciones y sus antiderivadas
x_vals = np.linspace(-5, 5, 400)  # Valores de x para la grfica

plt.figure(figsize=(12, len(funciones) * 4))

for i, f in enumerate(funciones, 1):
    # Convertir la funcin y la integral a funciones numricas usando lambdify
    f_num = lambdify(x, f, 'numpy')
    F_num = lambdify(x, integrales[f], 'numpy')

    # Graficar la funcin original
    plt.subplot(len(funciones), 2, 2 * i - 1)
    plt.plot(x_vals, f_num(x_vals), label=f'$f(x) = {latex(f)}$', color='blue')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title(f"Funcin Original: $f(x) = {latex(f)}$")
    plt.legend()
    plt.grid(True)

    # Graficar la integral indefinida
    plt.subplot(len(funciones), 2, 2 * i)
    plt.plot(x_vals, F_num(x_vals), label=f'$\int {latex(f)} \, dx = {latex(integrales[f])} + C$', color='red')
    plt.xlabel('x')
    plt.ylabel('F(x)')
    plt.title(f"Integral Indefinida: $\int {latex(f)} \, dx$")
    plt.legend()
    plt.grid(True)

plt.tight_layout()
plt.show()
```

# Integrales que Divergen o No Divergen

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Definir la funcin a integrar
def f(x):
    return np.exp(-x**2)  # Funcin gaussiana

# Calcular la integral desde - hasta 
integral, error = quad(f, -np.inf, np.inf)

# Verificar si la integral converge
if np.isfinite(integral) and error < 1e-4:
    print(f"La integral converge y su valor es: {integral:.4f}")
else:
    print("La integral diverge.")

# Graficar la funcin
x_vals = np.linspace(-5, 5, 400)
y_vals = f(x_vals)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label=r"$f(x) = e^{-x^2}$", color='blue')
plt.fill_between(x_vals, y_vals, alpha=0.2, color='cyan', label="rea bajo la curva")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Integral de $f(x) = e^{-x^2}$ en el intervalo $(-\infty, \infty)$")
plt.legend()
plt.grid(True)
plt.show()
```

# Mximos y Mnimos

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, diff, solve, lambdify, sympify, log, exp

# Solicitar al usuario que ingrese la funcin
funcion_str = input("Introduce la funcin en trminos de x (por ejemplo, 'x**3 - 3*x**2 + 4' o 'exp(x) - x*log(x)'): ")
x = symbols('x')
f = sympify(funcion_str)  # Convertir la cadena de texto en una expresin simblica

# Calcular la primera y segunda derivada
f_prime = diff(f, x)
f_double_prime = diff(f_prime, x)

# Encontrar los puntos crticos resolviendo f'(x) = 0
puntos_criticos = solve(f_prime, x)

# Determinar si cada punto crtico es un mximo, mnimo o punto de inflexin
extremos = []
for punto in puntos_criticos:
    valor_segunda_derivada = f_double_prime.subs(x, punto)
    if valor_segunda_derivada > 0:
        extremos.append((punto, "Mnimo"))
    elif valor_segunda_derivada < 0:
        extremos.append((punto, "Mximo"))
    else:
        extremos.append((punto, "Inconcluso"))

# Mostrar los puntos crticos y su clasificacin
for punto, tipo in extremos:
    print(f"Punto crtico en x = {punto}: {tipo}")

# Convertir la funcin a una funcin numrica para graficar
f_num = lambdify(x, f, 'numpy')

# Definir el rango de valores de x para la grfica (evitar logaritmos de valores no permitidos)
x_vals = np.linspace(-10, 10, 400)
x_vals = x_vals[x_vals > 0.1] if 'log' in funcion_str else x_vals  # Evita valores negativos en logaritmos

# Graficar la funcin y marcar los extremos
y_vals = f_num(x_vals)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label=f'$f(x) = {funcion_str}$', color='blue')

# Marcar los mximos y mnimos en la grfica
for punto, tipo in extremos:
    if punto.is_real:
        y_punto = f_num(float(punto))
        color = 'red' if tipo == "Mximo" else 'green'
        plt.plot(float(punto), y_punto, 'o', color=color, label=f'{tipo} en x = {punto}')

# Configurar la grfica
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title(f"Mximos y Mnimos de la Funcin $f(x) = {funcion_str}$")
plt.legend()
plt.grid(True)
plt.show()
```

# Grficos 3D

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, sympify, lambdify, log, exp
from mpl_toolkits.mplot3d import Axes3D

# Solicitar la funcin al usuario
funcion_str = input("Introduce la funcin en trminos de x e y (por ejemplo, 'x**2 + y**2 - 4*x - 2*y' o 'exp(x) - y*log(y)'): ")

# Definir las variables simblicas
x, y = symbols('x y')
# Convertir la cadena de texto en una expresin simblica
f = sympify(funcion_str)

# Convertir la funcin simblica a una funcin numrica para evaluarla en numpy
f_num = lambdify((x, y), f, 'numpy')

# Rango de valores para x e y
x_vals = np.linspace(-5, 5, 100)
y_vals = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x_vals, y_vals)

# Calcular Z usando la funcin numrica
# Agregar un manejo de errores para evitar valores no definidos (por ejemplo, log(0) o log(negativos))
try:
    Z = f_num(X, Y)
except ValueError as e:
    print(f"Error al evaluar la funcin: {e}")
    Z = np.zeros_like(X)  # En caso de error, usar una superficie plana

# Graficar la funcin en 3D
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.7)

# Configurar la grfica
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("f(x, y)")
plt.title(f"Grfica de la Funcin $f(x, y) = {funcion_str}$")
plt.show()
```

# Respuesta de Frecuencias de Filtros

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, cheby1, cheby2, ellip, bessel, freqz

# Definir el diseo del filtro
def diseo_filtro(tipo_filtro, fc, fs, orden=2, ripple_db=1, atenuacion_db=40):
    nyquist = 0.5 * fs  # Frecuencia de Nyquist
    normal_fc = fc / nyquist

    if tipo_filtro == 'butterworth':
        b, a = butter(orden, normal_fc, btype='low', analog=False)
    elif tipo_filtro == 'chebyshev1':
        b, a = cheby1(orden, ripple_db, normal_fc, btype='low', analog=False)
    elif tipo_filtro == 'chebyshev2':
        b, a = cheby2(orden, atenuacion_db, normal_fc, btype='low', analog=False)
    elif tipo_filtro == 'elliptic':
        b, a = ellip(orden, ripple_db, atenuacion_db, normal_fc, btype='low', analog=False)
    elif tipo_filtro == 'bessel':
        b, a = bessel(orden, normal_fc, btype='low', analog=False, norm='phase')
    else:
        raise ValueError("Tipo de filtro no soportado.")
    return b, a

# Definir las caractersticas del filtro
tipo_filtro = input("Elige el tipo de filtro ('butterworth', 'chebyshev1', 'chebyshev2', 'elliptic', 'bessel'): ")
fc = float(input("Introduce la frecuencia de corte (Hz): "))
fs = float(input("Introduce la frecuencia de muestreo (Hz): "))
orden = int(input("Introduce el orden del filtro (1 o 2): "))

# Parmetros adicionales para Chebyshev y Elptico
ripple_db = 1  # en dB (slo para Chebyshev tipo 1 y Elptico)
atenuacion_db = 40  # en dB (slo para Chebyshev tipo 2 y Elptico)

# Calcular los coeficientes del filtro
try:
    b, a = diseo_filtro(tipo_filtro, fc, fs, orden, ripple_db, atenuacion_db)
except ValueError as e:
    print(f"Error: {e}")
else:
    # Calcular la respuesta en frecuencia del filtro
    w, h = freqz(b, a, worN=8000, fs=fs)

    # Graficar la respuesta en frecuencia
    plt.figure(figsize=(12, 6))
    plt.plot(w, 20 * np.log10(abs(h)), 'b')
    plt.title(f"Respuesta en Frecuencia del Filtro {tipo_filtro.capitalize()} - Orden {orden}")
    plt.xlabel("Frecuencia (Hz)")
    plt.ylabel("Ganancia (dB)")
    plt.grid()
    plt.axvline(fc, color='red', linestyle='--', label=f'Frecuencia de Corte: {fc} Hz')
    plt.legend()
    plt.show()
```
